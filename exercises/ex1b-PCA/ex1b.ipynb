```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python

# Ex1
# load the iris dataset 
from calendar import c


iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(iris_df.head())

# Ex2
# Compute the variance of the features
var = iris_df.var()     
print(var)

# Ex3
# compute the covariance matrix
cov = iris_df.cov()
print(cov.keys())

# Covariance between sepal length and sepal width
cov.iloc[0,1]

# Ex4
sns.pairplot(iris_df)
```

```python
# Ex5
from matplotlib.pylab import f


data = iris_df - iris_df.mean(axis=0)   
data.cov()

# Ex6
values, vectors = np.linalg.eig(data.cov())
print(f"Eigenvalues: {values}")
print(f"Eigenvectors: {vectors}")
```

```python
# Ex7
v_norm = values / values.sum() * 100
plt.plot(v_norm)
plt.xlabel('Principal component')
plt.ylabel('Percent explained variance')
plt.ylim([0, 100])

plt.show()

# Ex8
pc_proj = vectors.T.dot(data.T)
sns.pairplot(pd.DataFrame(pc_proj.T, columns=['PC1', 'PC2', 'PC3', 'PC4']))
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(iris_df)
values_pca = pca.explained_variance_ratio_ 
exp_var_ratio = pca.explained_variance_ratio_ 
vectors_pca = pca.components_

print(f"Eigenvalues: {values_pca}")
print(f"Eigenvectors: {vectors_pca}")
print(f"Explained variance ratio: {exp_var_ratio}")

data_transform = pca.transform(iris_df)

sns.pairplot(pd.DataFrame(data_transform, columns=['PC1', 'PC2', 'PC3', 'PC4']))
```
